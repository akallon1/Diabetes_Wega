# **Data Cleaning**

```python
# Check for zeros in columns that should not have zeros
cols_of_interest = ['Glucose', 'BloodPressure', 'SkinThickness', 'Insulin', 'BMI']

# Check for zeros in the specified columns
missing_data = df[cols_of_interest] == 0

# Sum the number of zeros in each column
missing_summary = missing_data.sum()

# Calculate the percentage of zeros relative to the total number of rows
missing_percentage = missing_summary / len(df)

# Display the summary
print(f'Number of zeros:\n{missing_summary}\n')
print(f'Percentage of zeros:\n{missing_percentage}')
```

```python
# Visualize missing data
plt.figure(figsize=(10, 3))
sns.heatmap(missing_data,yticklabels = False, cbar=False, cmap='viridis')
plt.title('Heatmap of Missing Data')
plt.show()
```

```python
df_copy = df
```

## **Handling Missing Data**

```python
# Mean/Median imputation
df.loc[df['Glucose'] == 0, 'Glucose'] = df['Glucose'].mean().astype('int64')
df.loc[df['BloodPressure'] == 0, 'BloodPressure'] = df['BloodPressure'].mean().astype('int64')
df.loc[df['Insulin'] == 0, 'Insulin'] = df['Insulin'].median().astype('int64')
df.loc[df['SkinThickness'] == 0, 'SkinThickness'] = df['SkinThickness'].median().astype('int64')
df.loc[df['BMI'] == 0, 'BMI'] = df['BMI'].median().astype('float64')


# Check if the zeros have been imputed
df[['Glucose', 'BloodPressure', 'SkinThickness', 'Insulin', 'BMI']].describe()
```

```python
# Visualize missing data
plt.figure(figsize=(10, 3))
sns.heatmap(missing_data,yticklabels = False, cbar=False, cmap='viridis')
plt.title('Heatmap of Missing Data')
plt.show()
```
